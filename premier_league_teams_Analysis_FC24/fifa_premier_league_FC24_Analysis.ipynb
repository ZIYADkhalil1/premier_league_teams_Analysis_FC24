## 9. Defensive style and the attack style:

### creating 4 pie charts to find a relation between the defensive style and the attack style

```python
pos['def_style'].unique()
```

```python
forward['def_style'].unique()
```

```python
direct['def_style'].unique()
```

```python
balance['def_style'].unique()
```

```python
import matplotlib.pyplot as plt

# Assuming 'balance', 'direct', 'forward', and 'pos' are your DataFrames
dataframes = [balance, direct, forward, pos]
title = ['offensive chance creation is balanced', 'offensive chance creation is Direct passing', 'offensive chance creation is Forward runs','offensive chance creation is Possession ']
# Create a figure with subplots
fig, axs = plt.subplots(2, 2, figsize=(12, 10))
fig.suptitle('Distribution of Defensive Styles', fontsize=16)

# Flatten the axs array for easier indexing
axs = axs.flatten()

# Plot pie charts for each DataFrame
for i, df in enumerate(dataframes):
    def_style_counts = df['def_style'].value_counts()
    axs[i].pie(def_style_counts, labels=def_style_counts.index, autopct='%1.1f%%', startangle=90, colors=plt.cm.Set3.colors)
    axs[i].set_title(title[i])  # Use an alternative label

# Adjust layout
plt.tight_layout(rect=[0, 0, 1, 0.96])  # Leave space for the suptitle

# Show the pie charts
plt.show()
```
